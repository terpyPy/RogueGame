```python
def update(event_list, lookup):
    for pressed in event_list:
        lookup[pressed]['state'] = not lookup[pressed]['state']
        neighbor_update = lookup[pressed]['neighbors']
        for neighbor in neighbor_update:
            lookup[neighbor]['state'] = not lookup[neighbor]['state']
```

```python
# find factor pairs of a number
def factor_pairs(num):
    factors = np.arange(1, num+1)
    pairs = factors[num % factors == 0]
    pair_diffs = [(i, num//i) for i in pairs]
    pair_diffs.sort(key=lambda x: abs(x[0]-x[1]))
    return pair_diffs[0]
```

```python
# find largest square grid that fits in a rectangle
def largest_square_grid(num):
    factor1, factor2 = factor_pairs(num)
    if factor1 > factor2:
        factor1, factor2 = factor2, factor1
    return factor1, factor2
```

```python
def eight_neighbors_adj(matrix, rowNumber, colNumber):
    rows = len(matrix)
    cols = len(matrix[0])
    for rowAdd in range(-1, 2):
        newRow = rowNumber + rowAdd
        if 0 <= newRow < rows:
            for colAdd in range(-1, 2):
                newCol = colNumber + colAdd
                if 0 <= newCol < cols and (newRow != rowNumber or newCol != colNumber):
                    yield int(matrix[newRow][newCol])
    
# newer version of the adjacency algorithm that uses generators
# lets test it out
s = np.arange(0,16)
rows, cols = largest_square_grid(len(s))
s = s.reshape(rows, cols)
# calculate each element's 8 adjacent neighbors i.e. 1's neighbors are [2, 4, 5]
neighbors = []
for row in range(rows):
    for col in range(cols):
        print(list(eight_neighbors_adj(s, row, col)))
```

the adjacency algorithm is O(n^2) where n is the number of nodes in the graph, only computed once at the beginning.

Since we are using a linked list, the space complexity is O(n) where n is the number of nodes in the graph. 

The space complexity of the adjacency matrix is meaningless because its only used to generate the linked list, and is not used in the actual algorithm. 

We are trading space & initialization time complexity, for speed in the actual game.

```python
s = np.arange(0,16)
rows, cols = largest_square_grid(len(s))
s = s.reshape(rows, cols)
# calculate each element's 8 adjacent neighbors i.e. 1's neighbors are [2, 4, 5]
neighbors = []
for row in range(rows):
    for col in range(cols):
        # print(eight_neighbors_adj(s, row, col))
        neighbors.append(eight_neighbors_adj(s, row, col))
```
